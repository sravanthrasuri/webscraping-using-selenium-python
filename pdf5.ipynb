```python
import PyPDF2
import re

def extract_info_from_pdf(pdf_path):
    session_names = []
    chairs = []
    sub_sessions_and_speakers = []

    with open(pdf_path, 'rb') as file:
        pdf_reader = PyPDF2.PdfReader(file)
        num_pages = len(pdf_reader.pages)

        for page_number in range(num_pages):
            page = pdf_reader.pages[page_number]
            text = page.extract_text()

            # Extract session names
            session_names.extend(re.findall(r'\d{2}:\d{2} – \d{2}:\d{2} \/ (.*?)\n', text))

            # Extract chairs
            chairs.extend(re.findall(r'Moderators: (.*?)\n', text))

            # Extract sub-sessions and speakers
            sub_sessions_and_speakers.extend(re.findall(r'\d{2}:\d{2} (.*?)\n(.*?)(?=\d{2}:\d{2}|\n#LVL)', text, re.DOTALL))

    return session_names, chairs, sub_sessions_and_speakers

def print_info_separately(session_names, chairs, sub_sessions_and_speakers):
    # Print session names separately
    print("Session Names:")
    for session_name in session_names:
        print(session_name)

    # Print chairs separately
    print("\nChairs:")
    for chair in chairs:
        print(chair)

    # Print sub-sessions and speakers separately
    print("\nSub-Sessions and Speakers:")
    for sub_session, speakers in sub_sessions_and_speakers:
        print("Sub-Session:", sub_session.strip())
        print("Speakers:", speakers.strip())
        print()  # Add a blank line for separation

# Provide the path to your PDF file
pdf_path = r"/Users/sravanthrasuri/Downloads/ESGAR24-FinalProgramme-WEB_read_only-unlocked.pdf"
session_names, chairs, sub_sessions_and_speakers = extract_info_from_pdf(pdf_path)

# Print the extracted information separately
print_info_separately(session_names, chairs, sub_sessions_and_speakers)
```

```python
import PyPDF2
import re

def extract_info_from_pdf(pdf_path, start_page=None, end_page=None):
    session_names = []
    chairs = []
    sub_sessions = []
    speakers = []

    with open(pdf_path, 'rb') as file:
        pdf_reader = PyPDF2.PdfReader(file)

        if start_page is None:
            start_page = 0
        if end_page is None:
            end_page = len(pdf_reader.pages)

        for page_number in range(start_page, end_page):
            page = pdf_reader.pages[page_number]
            text = page.extract_text()

            # Extract session names
            session_names.extend(re.findall(r'\d{2}:\d{2} – \d{2}:\d{2} \/ (.*?)\n', text))

            # Extract chairs
            chairs.extend(re.findall(r'Moderators: (.*?)\n', text))

            # Preprocess text to remove unwanted patterns
            text = re.sub(r'\b(?:Monday|Tuesday|Wednesday|Thursday|Friday|Saturday|Sunday), (?:January|February|March|April|May|June|July|August|September|October|November|December) \d{1,2} \d{2}|ET \d+.*|\b\d{2}:\d{2}\s+SS\s+\d+\.\d+|#\w+\s+|ET \d+ Room \w+', '', text)
            text = re.sub(r'\b\d{2}:\d{2}\s+.*', '', text)  # Remove time before sentences

            # Extract sub-sessions and speakers
            sub_sessions.extend(re.findall(r'(?<=\n)(?![\d:]).*?(?=\n\d{2}:\d{2} – \d{2}:\d{2} \/ )', text, re.DOTALL))
            speakers.extend(re.findall(r'\d{2}:\d{2} (.*?)\n(.*?)(?=\d{2}:\d{2}|\n#LVL)', text, re.DOTALL))

    return session_names, chairs, sub_sessions, speakers

def print_info_separately(session_names, chairs, sub_sessions, speakers):
    # Print session names separately
    print("Session Names:")
    if session_names:
        for session_name in session_names:
            print(session_name)
    else:
        print(" ")

    # Print chairs separately
    print("\nChairs:")
    if chairs:
        for chair in chairs:
            print(chair)
    else:
        print(" ")

    # Print sub-sessions separately
    print("\nSub-Sessions:")
    if sub_sessions:
        for sub_session in sub_sessions:
            print(sub_session)
    else:
        print(" ")

    # Print speakers separately
    print("\nSpeakers:")
    if speakers:
        for speaker in speakers:
            print(speaker)
    else:
        print(" ")

# Provide the path to your PDF file
pdf_path = r"/Users/sravanthrasuri/Downloads/ESGAR24-FinalProgramme-WEB_read_only-unlocked.pdf"

# Extract information from specific pages of the PDF (e.g., from page 18 to page 88)
start_page = 18  # Change start page number as needed
end_page = 88  # Change end page number as needed
session_names, chairs, sub_sessions, speakers = extract_info_from_pdf(pdf_path, start_page, end_page)

# Print the extracted information separately
print_info_separately(session_names, chairs, sub_sessions, speakers)
```
